```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
X = df[['Literacy Rate', 'Rural Population', 'Urban Population', 'Total Population']]
y = df['GDP']
```

```python

model = LinearRegression()


model.fit(X, y)
```

```python

y_pred = model.predict(X)


plt.figure(figsize=(10, 6))
plt.scatter(y, y_pred)
plt.xlabel("Actual GDP")
plt.ylabel("Predicted GDP")
plt.title("Actual GDP vs. Predicted GDP")
plt.show()
```

```python


# Define the target variable (the one you want to predict)
target_variable = 'GDP'

# List of predictor variables
predictor_variables = ['Literacy Rate', 'GNI', 'Rural Population', 'Urban Population', 'Population Density', 'Total Population']

# Perform linear regression for each predictor variable
for predictor in predictor_variables:
    X = df[[predictor]]
    y = df[target_variable]

    # Create a linear regression model
    model = LinearRegression()

    # Fit the model to the data
    model.fit(X, y)

    # Make predictions
    y_pred = model.predict(X)

    # Visualize the results
    plt.figure(figsize=(6, 4))
    plt.scatter(X, y, label='Actual')
    plt.plot(X, y_pred, color='red', label='Predicted')
    plt.xlabel(predictor)
    plt.ylabel(target_variable)
    plt.title(f"{predictor} vs. {target_variable} Regression")
    plt.legend()
    plt.show()
```

```python
from sklearn.linear_model import LinearRegression

# List of dependent variables
dependent_variables = ['GDP', 'Literacy Rate', 'GNI', 'Rural Population', 'Urban Population', 'Total Population']

# Common predictor variables
predictor_variables = ['Literacy Rate', 'GNI', 'Rural Population', 'Urban Population', 'Population Density', 'Total Population']

# Create a dictionary to store R-squared values
r_squared_values = {}

# Perform regression for each dependent variable
for dependent_variable in dependent_variables:
    X = df[predictor_variables]
    y = df[dependent_variable]

    # Create a linear regression model
    model = LinearRegression()

    # Fit the model to the data
    model.fit(X, y)

    # Calculate R-squared
    r_squared = model.score(X, y)

    # Store R-squared value in the dictionary
    r_squared_values[dependent_variable] = r_squared

# Print R-squared values
for dependent_variable, r_squared in r_squared_values.items():
    print(f"R-squared for {dependent_variable}: {r_squared}")
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np

# List of dependent variables
dependent_variables = ['GDP', 'Literacy Rate', 'GNI', 'Rural Population', 'Urban Population', 'Total Population']

# Common predictor variables
predictor_variables = ['Literacy Rate']

# Create a dictionary to store Multiple R values
multiple_r_values = {}

# Perform regression for each dependent variable
for dependent_variable in dependent_variables:
    X = df[predictor_variables]
    y = df[dependent_variable]

    # Create a linear regression model
    model = LinearRegression()

    # Fit the model to the data
    model.fit(X, y)

    # Predicted values
    y_pred = model.predict(X)

    # Calculate Multiple R by squaring the Pearson correlation coefficient
    multiple_r = np.corrcoef(y, y_pred)[0, 1]

    # Store Multiple R value in the dictionary
    multiple_r_values[dependent_variable] = multiple_r

# Print Multiple R values
for dependent_variable, multiple_r in multiple_r_values.items():
    print(f"Multiple R for {dependent_variable}: {multiple_r:.4f}")
```

 "Interpretation"
- the regression analysis with 'Literacy Rate':
The multiple R for 'Literacy Rate' and 'GDP' is 0.2820, indicating a weak positive correlation between literacy rate and GDP.
This suggests that countries with higher literacy rates tend to have slightly higher GDPs.
-the regression analysis with 'GNI':
The multiple R for 'GNI' and 'GDP' is 0.9999, indicating an extremely strong positive correlation between Gross National Income (GNI) and GDP.
This suggests a near-perfect linear relationship between GNI and GDP, which is expected, as GNI is a major component of GDP.
-regression analysis with 'Total Population':
The multiple R for 'Total Population' and 'GDP' is 0.4896, indicating a moderate positive correlation.
This implies that larger populations are associated with higher GDPs to some extent.
-insights from the correlation matrix
 the positive correlation between 'GNI' and 'Urban Population' (0.6473), indicating that as GNI increases, urbanization tends to rise.
 the strong positive correlation between 'Urban Population' and 'Rural Population' (0.7897), suggesting that countries with higher urban populations also tend to have larger rural populations.
 the strong positive correlation between 'Rural Population' and 'Total Population' (0.9443), indicating that as the rural population increases, so does the total population.

Conclusions-

-There is a positive, albeit weak, correlation between literacy rate and GDP, suggesting that literacy can have a modest impact on economic performance.
-The relationship between GNI and GDP is exceptionally strong, underscoring the fundamental role of GNI in determining a country's economic output.
-The size of the total population shows a moderate positive correlation with GDP, indicating population size does play a role in economic performance.
-Urbanization is positively associated with GNI, reflecting the trend of higher income levels in urban areas.
-The strong correlation between rural and urban populations implies that changes in one often correspond with changes in the other.
